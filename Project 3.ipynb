```python
cat_to_dummies = [
'DistArea_ID'               
,'Medicine_Type'
,'SidEffect_Level'
,'Area_Type'
,'Area_City_Type'
,'Area_dist_level' 
]


simple_numeric = [
'Counterfeit_Weight'   
,'Active_Since'           
,'Medicine_MRP'
,'Availability_rating'
]
```

```python
mypipe=dt.DataPipe(cat_to_dummies=cat_to_dummies,simple_numeric=simple_numeric)
```

```python
mypipe.fit(df_train)
```

```python
X_train=mypipe.transform(df_train)
```

```python
X = X_train
y = df_train['Counterfeit_Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()

X_train= scaler.fit_transform(X_train)

X_test= scaler.transform(X_test)
```

```python
df_test=mypipe.transform(df_test)

# Create StandardScaler Object
scaler = StandardScaler()

# Perform Scaling
df_test= scaler.fit_transform(df_test)
```

```python
gbm = GradientBoostingRegressor(
    loss='absolute_error',  # ls stands for least squares regression (for continuous outputs)
     learning_rate=0.1,
    n_estimators=100,
    subsample=1.0,
    criterion='friedman_mse',  # mse or mae (mean squared error or mean absolute error)
    min_samples_split=2,
    min_samples_leaf=1,
    max_depth=3,
    random_state=42
)


param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

# Use GridSearchCV to search for the best parameters
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=50, scoring='neg_mean_absolute_error', verbose=2)
grid_search.fit(X_train, y_train)

# Print the best parameters and best MAE score
print("Best Parameters:", grid_search.best_params_)
print("Best MAE Score:", -grid_search.best_score_)  # Since we used 'neg_mean_absolute_error', we take the negative to get MAE

# Get the best model
best_gbm = grid_search.best_estimator_

# Train the model
#best_gbm.fit(X_train, y_train)

# Make predictions
predictions = best_gbm.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
print("Mean Absolute Error (MAE):", mae)

Score = 1-(mae/1660)
Score
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.6f} (std: {1:.6f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```
